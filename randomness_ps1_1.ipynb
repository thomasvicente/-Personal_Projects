```python
import random
import matplotlib.pyplot as plt

def plot_running_fraction_of_twos():
    rolls = [random.randint(1, 6) for _ in range(1000)] # Generate a pseudo-random sequence of 1000 simulated rolls
    running_fraction_of_twos = [0] # Initialize the running fraction of twos
    two_count = 0 # Initialize the count of twos
    for i in range(1, len(rolls) + 1):
        if rolls[i-1] == 2:
            two_count += 1
        running_fraction_of_twos.append(two_count/i)
    # Plot the running fraction of twos
    plt.plot(running_fraction_of_twos)
    plt.xlabel("Number of rolls")
    plt.ylabel("Running fraction of twos")
    plt.show()

plot_running_fraction_of_twos()
```
